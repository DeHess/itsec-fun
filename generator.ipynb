```python
import numpy as np

def find_generators(start, end, operation):
    nrRange = np.arange(start, end)
    z = end
    vorgekommen = []
    generators = []
    
    if operation == '+':
        current = 0
        for a in nrRange:
            print("===== a:", a, " =====")
            for i in range(z):
                current = current + a
                g = current % z
                print("i:", i+1, "current:", current, "Modulo:", g)
                if not (g in vorgekommen):
                    vorgekommen.append(g)
                
            if len(vorgekommen) == z:
                print("Alle Zahlen kommen vor:", a, "ist ein Generator!")
                generators.append(a)
            current = 0
            vorgekommen = []
    
    if operation == '*':
        current = 1
        for a in nrRange:
            print("===== a:", a, " =====")
            for i in range(z):
                current = current * a
                g = current % z
                print("i:", i+1, "current:", current, "Modulo:", g, flush='True')
                if not (g in vorgekommen):
                    vorgekommen.append(g)
                
            if len(vorgekommen) == z-1:
                print("Alle Zahlen kommen vor:", a, "ist ein Generator!")
                generators.append(a)
            current = 1
            vorgekommen = []
    
    print("Generatoren (a) aus Z(", z, operation ,"): ",  generators)
        
        
#Z7 ist mit 0
#Z7* ist ohne 0
    
find_generators(1, 7, '*')
```
